```python
import numpy as np
import statistics as st
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
arr = [34,25,46,25,48,39,36,29,42,37,28]
```

```python
# measure of central tendency
# mean, median, mode

print(np.mean(arr))
print(np.median(arr))
print(st.mode(arr))
```

```python
sns.boxplot(arr)
```

```python
# check outlier[lower - higher]

q1, q3 = np.percentile(arr,[25, 75])

IQR = q3 - q1
lower = q1 - 1.5*(IQR)
higher = q3 + 1.5*(IQR)
print(lower, higher)
```

```python
# measure of dispersion
# 1. variance
# 2. standard deviation

print(np.var(arr))
print(np.std(arr))
```

```python
# histograms

sns.histplot(arr, kde = True)
```
